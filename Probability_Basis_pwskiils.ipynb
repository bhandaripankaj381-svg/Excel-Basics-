Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

#Solution
Stratified Sampling should be used because it divides the population into strata (boys and girls) and ensures proper representation of each group in the sample, unlike Simple Random Sampling.

Question 10: Differentiate between discrete and continuous random variables with examples.
#Answer
A discrete random variable takes countable and distinct values. These values can be listed one by one. It is generally related to counting.
Example: Number of students in a class, number of defective items, number of heads in coin tosses.

A continuous random variable takes uncountable values within a given range. It is generally related to measurement, and any value between two numbers is possible.
Example: Height of students, weight, time, temperature.
