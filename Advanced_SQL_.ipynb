# Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?

-->A Common Table Expression (CTE) is a temporary result set that can be referenced within a SELECT, INSERT, UPDATE, or DELETE query. It is defined using the WITH clause.

CTEs improve SQL query readability by:

Breaking complex queries into smaller logical parts

Making queries easier to understand and maintain

Avoiding repeated subqueries


#Example
WITH HighSalaryEmployees AS (
    SELECT EmployeeID, Name, Salary
    FROM Employees
    WHERE Salary > 50000
)
SELECT * FROM HighSalaryEmployees;

## Q2. Why are some views updatable while others are read-only?

--> A view is updatable when the database can clearly determine how changes in the view should affect the underlying table.

# Updatable Views

A view is updatable when:

It is based on a single table

It does not use GROUP BY, DISTINCT, or aggregate functions

It does not contain joins or complex queries

# Read-Only Views

A view becomes read-only when:

It uses JOINs

It contains GROUP BY or aggregation

It uses DISTINCT or UNION

## Example
Updable View
CREATE VIEW vw_ProductPrice AS
SELECT ProductID, ProductName, Price
FROM Products;

You can update
UPDATE vw_ProductPrice
SET Price = 500
WHERE ProductID = 1;

# Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

Stored procedures provide several advantages:

Reusability – Write once and use multiple times.

Performance – Stored procedures are precompiled and execute faster.

Security – Users can execute procedures without direct access to tables.

Reduced Network Traffic – Only procedure calls are sent instead of full queries.

Maintainability – Logic is centralized in the database




# Q4. What is the purpose of triggers in a database?

->A trigger is a special type of stored procedure that automatically executes when a specific event occurs in a table (such as INSERT, UPDATE, or DELETE).

Triggers are used for:

Maintaining data integrity

Logging changes

Auditing database activity

Enforcing business rules

Example Use Case

When a product is deleted from the Products table, a trigger can automatically store that deleted record in an archive table for backup or auditing purposes.

# Q5. Explain the need for data modelling and normalization when designing a database.

->Data Modelling

Data modelling defines how data is structured and how different tables relate to each other.

Benefits:

Improves database structure

Makes relationships clear

Ensures efficient data storage

Normalization

Normalization is the process of organizing data to reduce redundancy and improve data integrity.

Benefits:

Eliminates duplicate data

Improves consistency

Makes updates easier

Saves storage space

Example: Separating Customers and Orders into different tables instead of storing everything in one table.

# Q6. Write a CTE to calculate total revenue for each product where revenue > 3000?

-->WITH ProductRevenue AS (
    SELECT
        ProductID,
        ProductName,
        Price,
        Quantity,
        (Price * Quantity) AS Revenue
    FROM Products
)

SELECT *
FROM ProductRevenue
WHERE Revenue > 3000;

# Q7. Create a view vw_CategorySummary ?

--> CREATE VIEW vw_CategorySummary AS
SELECT
    Category,
    COUNT(ProductID) AS TotalProducts,
    AVG(Price) AS AveragePrice
FROM Products
GROUP BY Category;

# Q10. Create an AFTER DELETE trigger to archive deleted products ?


--> Archieve table

CREATE TABLE ProductArchive (
    ProductID INT,
    ProductName VARCHAR(100),
    Category VARCHAR(100),
    Price DECIMAL(10,2),
    DeletedAt DATETIME
);


trigger
CREATE TRIGGER trg_AfterDeleteProducts
ON Products
AFTER DELETE
AS
BEGIN
    INSERT INTO ProductArchive
    (ProductID, ProductName, Category, Price, DeletedAt)
    
    SELECT
        ProductID,
        ProductName,
        Category,
        Price,
        GETDATE()
    FROM deleted;
END;
