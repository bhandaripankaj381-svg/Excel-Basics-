# Cell Refrencing

1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.
#Realtive cell refrecing

-->In relative cell referencing, the cell address changes automatically when the formula is copied to another cell. For example, if a formula uses cell A1 and we copy it down, it will change to A2, A3, etc. This type of reference is used when we want the formula to adjust according to position.

#Absoulete cell Refrencing
--->In absolute cell referencing, the cell address remains fixed even when the formula is copied. It uses the $ sign to lock the row and column, for example $A$1. This means the formula will always refer to the same cell. It is used when we want to use a constant value.

#Mixed Cell Refrecing

Mixed Cell Referencing:
Mixed cell referencing locks either the row or the column, not both.
Example:
$A1 → Column A is fixed, row changes
A$1 → Row 1 is fixed, column changes
It is used when only one part of the cell address needs to remain constant.
