#Question 1 Create a New Database and  Table for Employees

CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE);

## Question 9 What is the Difference Between a Primary Key and a Foreign Key?

-->Primary Key
1. Uniquely identifies each record
2. Cannot contain NULL values
3. One per table
4. Ensures entity integrity

-->Foregin key
1. Refers to primary key of another table
2. Can contain NULL values
3. Multiple foreign keys allowed
4. Ensures referential integrity

Question 10. What Are Constraints in SQL and Why Are They Used?

--> Constarints are rules applied to table Coulumns to ensure data accuracy adn integrity.
Common SQL Constraints:

PRIMARY KEY – Unique identifier, no NULL values

FOREIGN KEY – Links two tables

NOT NULL – Prevents NULL values

UNIQUE – Ensures unique values

CHECK – Applies condition on data

DEFAULT – Sets default value

## Why Constraints Are Used:

Prevent invalid data entry

Maintain data consistency

Enforce business rules

Improve database reliability
